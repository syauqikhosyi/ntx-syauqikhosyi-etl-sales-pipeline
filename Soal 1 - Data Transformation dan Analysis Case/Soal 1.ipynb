# Second Question:
* Detect any anomalies, such as a sharp decrease or increase in the number of transactions for a specific product.

ChatGPT Response Code & Improve Code:

```python
# ChatGPT Response Code
# Group by date and product_name, then sum transaction counts
daily_transactions = cleaned_data.groupby(['date', 'v2ProductName'])['transactions'].sum().reset_index()

# Display the result
daily_transactions.head()
```

```python
# ChatGPT Response Code & Improve Code
# Sort data by product and date
daily_transactions = daily_transactions.sort_values(by=['v2ProductName', 'date'])

# Calculate the difference in transaction counts for each product
daily_transactions['transaction_diff'] = daily_transactions.groupby('v2ProductName')['transactions'].diff()

# Display the result
daily_transactions.head()
```

```python
# ChatGPT Response Code
# Calculate mean and standard deviation of transaction differences
threshold_std = 2  # Ubah sesuai kebutuhan
stats = daily_transactions.groupby('v2ProductName')['transaction_diff'].agg(['mean', 'std']).reset_index()
stats['upper_bound'] = stats['mean'] + threshold_std * stats['std']
stats['lower_bound'] = stats['mean'] - threshold_std * stats['std']

# Merge bounds back to the main dataframe
daily_transactions = daily_transactions.merge(stats[['v2ProductName', 'upper_bound', 'lower_bound']], on='v2ProductName')

# Identify anomalies
daily_transactions['anomaly'] = (daily_transactions['transaction_diff'] > daily_transactions['upper_bound']) | \
                                 (daily_transactions['transaction_diff'] < daily_transactions['lower_bound'])

# Display anomalies
anomalies = daily_transactions[daily_transactions['anomaly'] == True]
anomalies
```

```python
# ChatGPT Response Code & Improve Code
# Plot transaction counts for a specific product with anomalies highlighted
product_name = "1 oz Hand Sanitizer"  # Change specific products
product_data = daily_transactions[daily_transactions['v2ProductName'] == product_name]

plt.figure(figsize=(12, 6))
plt.plot(product_data['date'], product_data['transactions'], label='Transaction Count')
plt.scatter(product_data[product_data['anomaly'] == True]['date'], 
            product_data[product_data['anomaly'] == True]['transactions'], 
            color='red', label='Anomaly')
plt.title(f"Transaction Count for {product_name} with Anomalies Highlighted")
plt.xlabel("Date")
plt.ylabel("Transaction Count")
plt.legend()
plt.grid(True)
plt.show()
```
